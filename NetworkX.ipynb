# 无向图生成

```python
import networkx as nx

G = nx.Graph()
G.add_node('s')
G.add_edge(2,3)
G.add_edge(3,2)
print (G.nodes())
print (G.edges())
print (G.number_of_edges())
```

# 加权图(网络)

```python
import networkx as nx

G = nx.Graph()
#添加0-1,1-2两条边,权重分别是3.0,7.5
G.add_weighted_edges_from([(0,1,3.0),(1,2,7.5)])
print (G.get_edge_data(1,2))
```

NetworkX提供了常用图论经典算法，可在http://networkx.lanl.gov/reference/algorithms.html 查看相应算法

# 统计指标计算

### 一、度、度分布 

NetworkX可以用来统计图中每个节点的度，并生成度分布序列

```python
import networkx as nx
G = nx.random_graphs.barabasi_albert_graph(1000,3) #生成一个 n=1000,m=3的BA无标度网络
print(G.degree(0))
print(G.degree())#返回所有节点的度
print(nx.degree_histogram(G))#返回图中所有节点的度分布序列
```

#### 绘图

```python
import matplotlib.pyplot as plt
degree = nx.degree_histogram(G)
x = range(len(degree))
y = [z / float(sum(degree)) for z in degree]
plt.loglog(x,y,color="blue",linewidth=2,label="distribution")
plt.legend()
plt.show()
```
