```python
import requests
from pyquery import PyQuery as pq

short_comment = url + "comments/hot?p=1"
print(short_comment)
html = requests.get(short_comment).text

doc = pq(html)
items = doc('#comment-list-wrapper #comments .comment-item').items()
for item in items:
    #book_url=item.find('a[href^=https]').attr('href')
    commenter= item.find('a[href^=https]').text()
    comment_main = item.find('.comment-content').text()
    comment = {
            'commenter': commenter,
            'comment-content': comment_main, 
            }
    print(comment)
```

```python
import re
import requests
book_review = url + "reviews"
html = requests.get(book_review).text
data_cid=re.compile(
                '<div xmlns:v="http://rdf.data-vocabulary.org/#" typeof="v:Review" data-cid="(.*?)">'
            )
re_data_cid = data_cid.findall(str(html))
for cid in re_data_cid:
    print("cid:{0} \n\n".format(cid))
    req = requests.get("https://book.douban.com/j/review/"+cid+"/full", timeout=6).json()
    print(req['html'].replace('</p>','\n').replace('<p>','').replace('<br>','\n').replace('&nbsp;',' '))
```
