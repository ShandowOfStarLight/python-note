# 正则表式

```python
import re
teststring="hello12334567xxbbcc124111"
result=re.findall(r'^hello(\d*).*?$',teststring)
print (result)
```

每次使用正则表达式的时候都是先把字符串都包含进去，然后用括号括住你想要的字段使用search获取第一个匹配或者使用findall获取所有匹配。

```python
import re
teststring="""helll1111333
vff dsfefsf
sfsf
"""
pattern=re.compile('^helll(.*?).*?',re.S)
result=re.match(pattern,teststring)
print (result)
```

编译模式串以达到复用的目的，使用re.compile(),存在换行符时加个re.S来匹配。

## 提取网站中文字符

```python

import requests
import re
content=requests.get(r'https://www.bannedbook.org/bnews/tag/%e4%b8%ad%e5%85%b1%e4%b8%ad%e5%a4%ae/').text
pattern=re.compile('[\u4e00-\u9fa5]',re.S)
results=re.findall(pattern,str(content))

with open("chinese.txt","a", encoding="utf-8") as input_f:
    input_f.write(''.join(results))
```

# 豆瓣ebook爬取

```python
import re
import requests
url = "https://read.douban.com/ebook/21926849/"
page = requests.get(url).text
print (page)
book_detail_pattern=re.compile(r'<h1 itemprop="name" class="article-title">(.*?)</h1>.*?<a class="author-item" .*?">(.*?)</a></span></p>.*?<p class=""><span class="label">字数</span><span class="labeled-text">(.*?)</span></p>',re.S)
book_detail = re.findall(book_detail_pattern,page)
print (book_detail)
```

## 标签爬取

```python
import requests
import re
tag_url = "https://book.douban.com/tag/"
tag_page = requests.get(tag_url).text
pattern=re.compile('<td><a href="/tag/.*?">(.*?)</a><b>.*?</b></td>',re.S)
tags = re.findall(pattern,tag_page)
newstr = ""
for tag in tags:
    newstr= newstr+tag+"、"
```
